```python
import numpy as np
import pandas as pd
import re

# Stopwords tiếng Anh cơ bản
stopwords = set("""
a an the and or in on with to of at for from by about as is are was were be been being
has have had do does did will would can could should this that these those so such
""".split())
```

```python
def preprocess(text):
    if not isinstance(text, str):
        return []
    text = text.lower()
    text = re.sub(r'[^a-z0-9\s]', '', text)
    words = text.split()
    return [word for word in words if word not in stopwords]

# Áp dụng tiền xử lý lên cột 'tags'
new_data['tokens'] = new_data['tags'].apply(preprocess)
```

```python
# Tạo từ điển từ tất cả các tokens
vocab = sorted(set(word for tokens in new_data['tokens'] for word in tokens))
word_to_index = {word: idx for idx, word in enumerate(vocab)}
vocab_size = len(vocab)

# Vector hóa: mỗi phim là một vector đếm từ (Bag of Words)
def vectorize(tokens):
    vec = np.zeros(vocab_size, dtype=np.float32)
    for word in tokens:
        if word in word_to_index:
            vec[word_to_index[word]] += 1
    return vec

# Tạo ma trận vector cho tất cả phim
vectors = np.stack(new_data['tokens'].apply(vectorize).values)
```

```python
# TF: tần suất xuất hiện của từ trong phim
def compute_tf(tokens):
    tf = np.zeros(vocab_size, dtype=np.float32)
    for word in tokens:
        if word in word_to_index:
            tf[word_to_index[word]] += 1
    tf /= len(tokens) if tokens else 1
    return tf

tf_vectors = np.stack(new_data['tokens'].apply(compute_tf).values)
```

```python
def recommend(movie_title, top_k=5):
    if movie_title not in new_data['title'].values:
        print("❌ Không tìm thấy phim:", movie_title)
        return

    idx = new_data[new_data['title'] == movie_title].index[0]
    scores = similarity_matrix[idx]

    # Loại bỏ chính nó (chỉ lấy các phim khác)
    similar_idxs = scores.argsort()[::-1]  # Sắp xếp giảm dần
    top_similar = [(i, scores[i]) for i in similar_idxs if i != idx][:top_k]

    print(f"\n🎬 Gợi ý phim tương tự với: '{movie_title}'\n")
    for i, score in top_similar:
        print(f"→ {new_data.iloc[i]['title']}")
```
