```python
import numpy as np
import pandas as pd
import re

# Stopwords ti·∫øng Anh c∆° b·∫£n
stopwords = set("""
a an the and or in on with to of at for from by about as is are was were be been being
has have had do does did will would can could should this that these those so such
""".split())
```

```python
def preprocess(text):
    if not isinstance(text, str):
        return []
    text = text.lower()
    text = re.sub(r'[^a-z0-9\s]', '', text)
    words = text.split()
    return [word for word in words if word not in stopwords]

# √Åp d·ª•ng ti·ªÅn x·ª≠ l√Ω l√™n c·ªôt 'tags'
new_data['tokens'] = new_data['tags'].apply(preprocess)
```

```python
# T·∫°o t·ª´ ƒëi·ªÉn t·ª´ t·∫•t c·∫£ c√°c tokens
vocab = sorted(set(word for tokens in new_data['tokens'] for word in tokens))
word_to_index = {word: idx for idx, word in enumerate(vocab)}
vocab_size = len(vocab)

# Vector h√≥a: m·ªói phim l√† m·ªôt vector ƒë·∫øm t·ª´ (Bag of Words)
def vectorize(tokens):
    vec = np.zeros(vocab_size, dtype=np.float32)
    for word in tokens:
        if word in word_to_index:
            vec[word_to_index[word]] += 1
    return vec

# T·∫°o ma tr·∫≠n vector cho t·∫•t c·∫£ phim
vectors = np.stack(new_data['tokens'].apply(vectorize).values)
```

```python
# TF: t·∫ßn su·∫•t xu·∫•t hi·ªán c·ªßa t·ª´ trong phim
def compute_tf(tokens):
    tf = np.zeros(vocab_size, dtype=np.float32)
    for word in tokens:
        if word in word_to_index:
            tf[word_to_index[word]] += 1
    tf /= len(tokens) if tokens else 1
    return tf

tf_vectors = np.stack(new_data['tokens'].apply(compute_tf).values)
```
